```python
import numpy as np

blade_volume_mm3 = 31.923
blade_volume_m3 = blade_volume_mm3 * 1e-9
density = 1140 # kg/m3 Gray V4 Resin

mass_per_blade = blade_volume_m3 * density
blade_number = 60
total_mass = mass_per_blade * blade_number

edge_diameter = 59 * 1e-3
counterbalance_height = 4 * 1e-3

equivalent_inner_diameter = np.sqrt( edge_diameter**2 - (4 * total_mass) / (np.pi * density * counterbalance_height))

print(f"Equivalent inner diameter: {equivalent_inner_diameter*1e3:.2f} mm")
```
