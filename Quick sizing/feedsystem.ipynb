```python
import numpy as np

tank_volume = 3.78 * 0.001  # L to m^3
initial_pressure = 90e5  # bar to Pa
regulated_pressure = 10e5  # bar to Pa

n2_R = 296.8  # J/(kgÂ·K)
n2_T0 = 293.15  # K

initial_density = initial_pressure / (n2_R * n2_T0)
regulated_density = regulated_pressure / (n2_R * n2_T0) # Assuming isothermal expansion

total_mass = initial_density * tank_volume  # kg

mass_flow = 0.047  # kg/s

discharge_time = total_mass / mass_flow - (tank_volume * regulated_density) / mass_flow  # s

print(f"Discharge time: {discharge_time:.2f} seconds")


```
