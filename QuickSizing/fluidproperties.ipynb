```python
from pyfluids import Fluid, FluidsList, Input

air = Fluid(FluidsList.Air).with_state(Input.temperature(182.9), Input.pressure(1.1e5))
# air = Fluid(FluidsList.Air).with_state(Input.temperature(298.15), Input.pressure(101325))
print (f'Temperature = {air.temperature:.2f} K')
print(f'Pressure = {air.pressure:.2f} Pa')
print(f'Density = {air.density:.4f} kg/m³')
print(f'Viscosity = {air.kinematic_viscosity:.7f} Pa·s')
print(f'Thermal Conductivity = {air.conductivity:.5f} W/(m·K)')
print(f'Pr = {air.prandtl:.5f}')
print(f'Cp = {air.specific_heat:.2f} J/(kg·K)')
```
