```python
import numpy as np

tank_volume = 3.78 * 0.001 * 2  # L to m^3
initial_pressure = 90e5  # bar to Pa
regulated_pressure = 7e5  # bar to Pa

n2_R = 296.8  # J/(kgÂ·K)
n2_T0 = 293.15  # K

initial_density = initial_pressure / (n2_R * n2_T0)
regulated_density = regulated_pressure / (n2_R * n2_T0) # Assuming isothermal expansion

total_mass = initial_density * tank_volume  # kg

mass_flow = 0.024  # kg/s

discharge_time = total_mass / mass_flow - (tank_volume * regulated_density) / mass_flow  # s

print(f"Discharge time: {discharge_time:.2f} seconds")

airflow_rate = 50 / 1000 / 60  # L/min to m^3/s
standard_density = 1.225  # kg/m^3
mass_flow_air = airflow_rate * standard_density  # kg/s
charge_time = total_mass / mass_flow_air  # s
print(f"Charge time: {charge_time:.2f} seconds")
```
